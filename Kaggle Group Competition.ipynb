```python
import numpy as np #for linear algebra
import pandas as pd # CSV file I/O 

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#Reading from data files
train = pd.read_csv('/kaggle/input/tcd-ml-comp-201920-income-pred-group/tcd-ml-1920-group-income-train.csv') 
test = pd.read_csv('/kaggle/input/tcd-ml-comp-201920-income-pred-group/tcd-ml-1920-group-income-test.csv')
```

```python
# Checking NULL values
train.isnull().sum()
```

```python
# Checing the mean, median, mode and all other relavent form of data per coulmn to analyse which data to use in filling the NULL Values.
# We did this to reduce the time of execution, for example if we write fillna(df[column_name].mean()) for each line it will take a lot more time for each column. 
train.describe(include="all")
```

```python
#filling the null values
null_dict = {'Year of Record': 1979.0,
             'Satisfation with employer': 'Average',
             'Gender': 'male',
             'Profession': 'payment analyst',
             'University Degree': 0,
             'Hair Color': 'Black',
             'Country': 'Honduras'
}
# Merging both the dataset so that we can apply operations at a same time
data = pd.concat([train,test],ignore_index=True)
for i in null_dict:
    data[i] = data[i].fillna(null_dict[i])
```

```python
# Checking if all the NULL values are removed or not.
data.isnull().sum()
```

```python
data.head(20)
```

```python
data.dtypes
```

```python
# Searching for categorical features so that we can apply Encoding on the same.
obj_cols = (data.dtypes[data.dtypes == 'object'].index).tolist()
obj_cols
```

```python
# Applying Label Encoding on categorical features
from sklearn.preprocessing import LabelEncoder
for i in obj_cols:
    label = LabelEncoder()
    label.fit(data[i].unique().astype(str))
    data[i] = label.transform(data[i].astype(str))
```

```python
# Spliting Data into test and train
from sklearn.model_selection import train_test_split
features =  list(set(data) - set(['Total Yearly Income [EUR]','Instances']))
X_train = data[features].iloc[:1048573]
Y_train = data['Total Yearly Income [EUR]'].iloc[:1048573]
X_test = data[features].iloc[1048574:]

# Dividing Train  data into test and train for finding the mean absolute error
x_train,x_val,y_train,y_val = train_test_split(X_train,Y_train,test_size=0.2,random_state=193322)
```

```python
import lightgbm as lgb
params = {
          'max_depth': 20,
          'learning_rate': 0.001,
          "boosting": "gbdt",
          "bagging_seed": 11,
          "metric": 'mse',
          "verbosity": -1,
         }
trn_data = lgb.Dataset(x_train, label=y_train)
val_data = lgb.Dataset(x_val, label=y_val)

clf = lgb.train(params, trn_data, 100000, valid_sets = [trn_data, val_data], verbose_eval=1000, early_stopping_rounds=500)

# Predicting the test data
pre_test_lgb = clf.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
# Predicting the remaining 20% of data from train to check the error rate.
predict = clf.predict(x_val)
val_mae = mean_absolute_error(y_val,predict)
```

```python
# Merging data of Instance and Prediction i.e. income.
X_instance = data['Instance'].iloc[1048574:]
sub_df = pd.DataFrame({'Instance':X_instance, 'Total Yearly Income [EUR]':(pre_test_lgb)})
sub_df.head()
```
